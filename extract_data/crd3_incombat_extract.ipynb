```python
special_tokens = [
        "<|begin_of_text|>",
        "<|end_of_text|>",
        "<|reserved_special_token_0|>",
        "<|reserved_special_token_1|>",
        "<|reserved_special_token_2|>",
        "<|reserved_special_token_3|>",
        "<|start_header_id|>",
        "<|end_header_id|>",
        "<|reserved_special_token_4|>",
        "<|eot_id|>",  # end of turn
    ]
```

```python
class InCombatDataset(Dataset):
    def __init__(self, df, labeled_df, tokenizer, max_len=512):
        self.df = df
        self.labeled_df = labeled_df
        self.max_len = max_len
        self.win_size = 50
        self.tokenizer = tokenizer

    def __len__(self):
        return len(self.labeled_df)

    def __getitem__(self, idx):
        df_idx = self.labeled_df.iloc[idx]['index'] - 1
        back_ = self.df.iloc[df_idx - self.win_size:df_idx]
        front_ = self.df.iloc[df_idx + 1:df_idx + self.win_size + 1]
        current_row = self.df.iloc[df_idx]
        context = []

        # for i in range(self.win_size):
        #     dialog = back_.iloc[i]['speaker'] + ': ' + back_.iloc[i]['utterance']
        #     context.append(dialog)
        #     context.append(self.tokenizer.sep_token)

        # prepare context for llama3 8b instruct
        context.append("<|start_header_id|>" + current_row['speaker'] + '<|end_header_id|>' + "\n\n"+ current_row['utterance'] + '<|eot_id|>')
        for i in range(self.win_size):
            dialog = "<|start_header_id|>" + front_.iloc[i]['speaker'] + '<|end_header_id|>' + "\n\n"+ front_.iloc[i]['utterance'] + '<|eot_id|>'
            context.append(dialog)
        context = ' '.join(context)
        # utterances = tokenizer(context, truncation=True, padding=True, return_tensors="pt", max_length=self.max_len)

        # label = self.labeled_df.iloc[idx]['in_combat']
        label_statement = 'IN_COMBAT' if self.labeled_df.iloc[idx]['in_combat'] == True else 'NOT_IN_COMBAT'
        label = '<|start_header_id|>' + 'system' + '<|end_header_id|>' + "\n\n" + label_statement + '<|eot_id|>'


        return context, label
```

```python
def test_formatter(df, idx, win_size=50):
    df_idx = idx - 1
    current_row = df.iloc[df_idx]
    front_ = df.iloc[df_idx + 1:df_idx + win_size + 1]
    context = []
    context.append("<|start_header_id|>" + current_row['speaker'] + '<|end_header_id|>' + "\n\n"+ current_row['utterance'] + '<|eot_id|>')
    for i in range(win_size):
        dialog = "<|start_header_id|>" + front_.iloc[i]['speaker'] + '<|end_header_id|>' + "\n\n"+ front_.iloc[i]['utterance'] + '<|eot_id|>'
        context.append(dialog)
    context = ' '.join(context)

    return context
```

```python
from transformers import Trainer, TrainingArguments
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForSequenceClassification

from transformers import DataCollatorWithPadding
from datasets import Dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, f1_score, precision_score, recall_score
```

```python
dataset = Dataset.from_pandas(data_to_train)
```

```python
dataset[1]
```

```python
dataset

dataset = dataset.map(lambda x: {'labels': 1 if x['label'] == True else 0}, remove_columns=['label'])
```

```python
dataset[0]
```

```python
# train model

data_to_model = dataset.train_test_split(test_size=0.25, shuffle=True, seed=42)

```

```python
data_to_model
def preprocess_function(examples):
    return tokenizer(examples["context"], truncation=True, padding=True, max_length=512)


tokenized_dataset = data_to_model.map(preprocess_function, batched=True)
tokenized_dataset
```

```python
import evaluate

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

accuracy = evaluate.load("accuracy")

label2id = {'True': 1, 'False': 0}
id2label = {v: k for k, v in label2id.items()}
num_labels = len(label2id)

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return accuracy.compute(predictions=predictions, references=labels)
```

```python
label2id, id2label, num_labels
```

```python

model = AutoModelForSequenceClassification.from_pretrained(
    model_name, num_labels=2, id2label=id2label, label2id=label2id
)
```

```python
training_args = TrainingArguments(
    output_dir="combat_model_manual_win_14",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    push_to_hub=True,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    processing_class=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

trainer.train()
# load the model from the checkpoint
# model = AutoModelForSequenceClassification.from_pretrained("ICOOC_model/checkpoint-10938")
```

```python
model = AutoModelForSequenceClassification.from_pretrained("combat_model_manual_win_14/checkpoint-230")
```

```python
def infer(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    inputs = {k: v.to(model.device) for k, v in inputs.items()}
    with torch.no_grad():
        logits = model(**inputs).logits
    predicted_class_id = logits.argmax().item()
    predicted_class = id2label[predicted_class_id]
    return predicted_class
```

```python
from tqdm import tqdm
win_size = 14
def infer_wrapper(df):
    for i in tqdm(range(df.shape[0]), desc="Inference"):
    # for i in range(20):
        index = df.iloc[i].name
        if i < win_size:
            pred = '-'
            # continue
        else :
            back_ = df.iloc[i - win_size:i]
            current_row = df.iloc[i]
            context = []
            for j in range(win_size):
                dialog = str(back_.iloc[j]['speaker']) + ': ' + str(back_.iloc[j]['utterance'])
                context.append(dialog)
                context.append(tokenizer.sep_token)
            context.append("[CURRENT] : " + str(current_row['speaker']) + ': ' + str(current_row['utterance']))
            context = ' '.join(context)

            pred = infer(context)
            if pred == 'False':
                pred = '-'
        # print(i,  index, pred)
        
        df.at[index, 'in_combat'] = pred
    
    return df
```
